```python
# all scrapped data combiner for all states csv to data frame states in it w

import pandas as pd
import numpy as np
```

```python
#csv to dataframe
df_buses_1AP = pd.read_csv(r"D:\Scraped_data\dfDATA_AP.csv")
df_buses_2CTU = pd.read_csv(r"D:\Scraped_data\dfDATA_CTU.csv")
df_buses_3HP = pd.read_csv(r"D:\Scraped_data\dfDATA_HP.csv")
df_buses_4KE = pd.read_csv(r"D:\Scraped_data\dfDATA_KE.csv")
df_buses_5PJ = pd.read_csv(r"D:\Scraped_data\dfDATA_PJ.csv")
df_buses_6RS = pd.read_csv(r"D:\Scraped_data\dfDATA_RS.csv")
df_buses_7TG = pd.read_csv(r"D:\Scraped_data\dfDATA_TG.csv")
df_buses_8UP = pd.read_csv(r"D:\Scraped_data\dfDATA_UP.csv")
df_buses_9WB1 = pd.read_csv(r"D:\Scraped_data\dfDATA_WB1SBSTC.csv")
df_buses_10WB2 = pd.read_csv(r"D:\Scraped_data\dfDATA_WB2WBTC.csv")
df_buses_11WB3 = pd.read_csv(r"D:\Scraped_data\dfDATA_WB3NBSTC.csv")
df_buses_12WB4 = pd.read_csv(r"D:\Scraped_data\dfDATA_WB4WBSTC.csv")
df_buses_13GOA = pd.read_csv(r"D:\Scraped_data\dfDATA_GOA.csv")
df_buses_14BIHAR = pd.read_csv(r"D:\Scraped_data\dfDATA_BIHAR.csv")
df_buses_15ASSAM1 = pd.read_csv(r"D:\Scraped_data\dfDATA_ASSAM.csv")
df_buses_16ASSAM2 = pd.read_csv(r"D:\Scraped_data\dfDATA_ASSAMkaac.csv")
df_buses_17JK = pd.read_csv(r"D:\Scraped_data\dfDATA_jkf.csv")

Final_df=pd.concat([df_buses_1AP,df_buses_2CTU,df_buses_3HP,df_buses_4KE,df_buses_5PJ,df_buses_6RS,
                    df_buses_7TG,df_buses_8UP,df_buses_9WB1,df_buses_10WB2,df_buses_11WB3,df_buses_12WB4,df_buses_13GOA,
                    df_buses_14BIHAR,df_buses_15ASSAM1,df_buses_16ASSAM2,df_buses_17JK])

Final_df
```

```python
Final_df.info()
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(480) 
```

```python
Final_df
```

```python
print(Final_df.columns)
```

```python
# Convert "New" to "4.0" as a string
Final_df["Star Rating"] = Final_df["Star Rating"].str.replace("New", "4.0")

# Strip any leading/trailing spaces
Final_df["Star Rating"] = Final_df["Star Rating"].str.strip()

# Split and take the first part
Final_df["Star Rating"] = Final_df["Star Rating"].str.split().str[0]

# Convert to numeric, coercing errors to NaN
Final_df["Star Rating"] = pd.to_numeric(Final_df["Star Rating"], errors='coerce')

# Fill NaN values with 3.0
Final_df["Star Rating"].fillna(3.0, inplace=True)
```

```python
Final_df
```

```python
# info after the data type change
Final_df.info()
```

```python
#only if you want to exclude buse with price more than 7000
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
Final_df
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
Final_df
```

```python
#convert the combined dataframe to csv with all state bus details    
path=r"D:\Scraped_data\dfFinal_all.csv"
Final_df.to_csv(path,index=False)
```
